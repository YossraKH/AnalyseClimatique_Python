```python
import pandas as pd
import numpy as np
```

```python
file_path = 'C:/Users/yossr/Analyse_climatique/Tunisia_Weather.csv'
Weather_data = pd.read_csv(file_path,encoding='utf-8')
print(Weather_data)
```

```python
Weather_data.info()
```

```python
Weather_data.drop_duplicates(inplace=True)
print(Weather_data.isnull().any(axis=1))
```

```python
print(Weather_data.head())
Weather_data.drop(columns=['T2MDEW','WD10M','WS50M','WD50M','ALLSKY_SFC_UV_INDEX'],inplace=True)
```

```python
#renommer les colonnes
new_names={
    'YEAR' : 'ANNEE',
    'MO':'MOIS',
    'DY' : 'JOUR',
    'ALLSKY_SFC_SW_DWN' : 'LUM_SOL_ATTEINT_SURFACE',
    'ALLSKY_KT':'INDICE_CIEL_CLAIRE',
    'ALLSKY_SFC_LW_DWN' : 'INFRAROUGE_NUAGE',
    'ALLSKY_SFC_PAR_TOT':'LUM_PHOTOSYNTHESE_NUAGE',
    'CLRSKY_SFC_PAR_TOT' : 'LUM_PHOTOSYNTHESE_CLAIRE',
    'ALLSKY_SFC_UVA':'INDICE_RAYONS_UVA',
    'ALLSKY_SFC_UVB':'INDICE_RAYONS_UVB',
    'T2M':'T2M',
    'T2M_MAX':'T2M_MAX',
    'T2M_MIN':'T2M_MIN',
    'TS':'T_SURFACE',
    'WS10M':'VITESSE_VENT_10M',
    'PRECTOTCORR':'PRECIPITATION',
    'PS': 'PRESSION',
    'RH2M':'HUMIDITE_2M'
}
Weather_data.rename(columns=new_names, inplace=True)
```

```python
print(Weather_data.head())
```

```python
Weather_data= Weather_data.drop(Weather_data[Weather_data['ANNEE'] < 2001].index)
Weather_data= Weather_data.drop(Weather_data[Weather_data['ANNEE'] > 2022].index)
Weather_data
```

```python
Weather_data = Weather_data.astype({'ANNEE': 'string', 'MOIS': 'string', 'JOUR':'string'})
Weather_data.info()
```

```python
Weather_data['DATE'] = Weather_data.apply(lambda row: f"{row['JOUR']}-{row['MOIS']}-{row['ANNEE']}", axis=1)
Weather_data
```

```python
Weather_data['DATE'] = pd.to_datetime(Weather_data['DATE'], format='%d-%m-%Y')
Weather_data.info()
```

```python
ordre_columns = ['ANNEE','MOIS','JOUR','DATE','T2M','T2M_MAX','T2M_MIN','T_SURFACE','LUM_SOL_ATTEINT_SURFACE','INDICE_CIEL_CLAIRE','INFRAROUGE_NUAGE',
                 'LUM_PHOTOSYNTHESE_NUAGE','LUM_PHOTOSYNTHESE_CLAIRE', 'INDICE_RAYONS_UVA','INDICE_RAYONS_UVB','VITESSE_VENT_10M','PRESSION',
                   'PRECIPITATION', 'HUMIDITE_2M']
Weather_data = Weather_data[ordre_columns]
Weather_data.info()
```

```python
output_file = 'C:/Users/yossr/Desktop/Clima_pipeline/Weather_output.csv'
Weather_data.to_csv(output_file,index=False,encoding='utf-8')
print('Les données sont enregistrées')
```
